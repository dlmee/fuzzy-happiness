```python
import matplotlib.pyplot as plt
import json

def plot_histogram(data_dict):
    aggregates = {}
    for v in data_dict.values():
        count = len(v)
        if count not in aggregates:
            aggregates[count] = 1
        else:
            aggregates[count] += 1

    categories = list(aggregates.keys())
    counts = [aggregates[k] for k in categories]
    print(categories)
    print(counts)
    plt.figure(figsize=(12, 8))  # Larger figure size
    plt.bar(categories, counts, color='blue')

    plt.xlabel('Number of Tokens per Stem')
    plt.ylabel('Frequency of Stems')
    plt.title('Histogram of Token Count Frequencies')
    plt.xticks(rotation=90)  # Rotate labels to prevent overlap
    plt.yscale('log')  # Use logarithmic scale for the y-axis

    plt.xlim(0, 200)  # Adjust x-axis limits as needed
    plt.tight_layout()
    plt.show()

# Load data from JSON and plot
with open("../side_hustle.json", "r") as inj:
    result = json.load(inj)
plot_histogram(result)
```
