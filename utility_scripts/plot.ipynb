```python
import json
import numpy as np

# Function to load word embeddings from a JSON file
def load_word_embeddings(json_file):
    with open(json_file, 'r') as inj:
        return json.load(inj)

# Function to calculate cosine similarity between two words
def cosine_similarity(word1, word2, embeddings):
    if word1 not in embeddings or word2 not in embeddings:
        return "Word not found in embeddings"
    else:
        embedding1 = embeddings[word1]
        embedding2 = embeddings[word2]
        dot_product = np.dot(embedding1, embedding2)
        norm1 = np.linalg.norm(embedding1)
        norm2 = np.linalg.norm(embedding2)
        similarity = dot_product / (norm1 * norm2)
        return similarity

# Example usage
json_file = '../data/w2vnep.json'
word_embeddings = load_word_embeddings(json_file)
```

```python
# Specify words to compare
cpair = ["", ""]
cpair = ["", ""]


# Calculate and print cosine similarity
similarity = cosine_similarity(cpair[0], cpair[1], word_embeddings)
print(f"Cosine similarity between '{cpair[0]}' and '{cpair[1]}': {similarity}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import json

def plot_word_length_distribution(data):
    word_length_counts = {}
    for stem, words in data.items():
        for word in words:
            length = len(word)
            if length not in word_length_counts:
                word_length_counts[length] = 0
            word_length_counts[length] += 1
    
    sorted_lengths = sorted(word_length_counts.keys())
    counts = [word_length_counts[length] for length in sorted_lengths]

    plt.figure(figsize=(10, 6))
    plt.bar(sorted_lengths, counts, align='center', alpha=0.7)
    plt.xlabel('Word Length')
    plt.ylabel('Log Count')
    plt.title('Log Distribution of Word Counts by Word Length')
    plt.yscale('log')
    plt.grid(True, which="both", ls="-")
    plt.show()

# Sample data
with open("../new_dictionary.json", "r") as inj:
    data = json.load(inj)

del data['**length**']

plot_word_length_distribution(data)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import json

def plot_stem_count_distribution(data):
    stem_word_counts = {}
    for stem, words in data.items():
        word_count = len(words['forms'])
        if word_count > 50: continue
        if word_count not in stem_word_counts:
            stem_word_counts[word_count] = 0
        stem_word_counts[word_count] += 1
    
    sorted_counts = sorted(stem_word_counts.keys())
    counts = [stem_word_counts[count] for count in sorted_counts]

    plt.figure(figsize=(10, 6))
    plt.bar(sorted_counts, counts, align='center', alpha=0.7)
    plt.xlabel('Number of Words per Stem')
    plt.ylabel('Log Count')
    plt.title('Log Distribution of Stem Counts by Number of Words per Stem')
    plt.yscale('log')
    plt.grid(True, which="both", ls="-")
    plt.show()

# Sample data
with open("../swahili/formatted_llm_response_ALL.json", "r") as inj:
    data = json.load(inj)

#del data['**length**']

plot_stem_count_distribution(data)
```
